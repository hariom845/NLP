# Spam Ham classification using Bow and TFIDF with naive-bayes model

1. Preprocessing and cleaning
2. Train-Test Split
3. BOW and TF-IDF (Sentences--->vectors){preventing data leakage}
4. Train our models

```python
import pandas as pd
messages=pd.read_csv('SMSSpamcollection.txt',sep='\t',names=['label','message'])
```

```python
messages
```

```python
# Data Cleaning and preprocessing
import nltk
import re
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
corpus=[]
for i in range(0, len(messages)):
    review=re.sub('[^a-zA-z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

# Creating Bag of Words

- independent-dependent features

```python
X=corpus
y=pd.get_dummies(messages['label']).astype(int)
y=y.iloc[:,1].values
X,y # here is two output feature so we can take any one as a output feature
```

```python
# train-test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20, random_state=42)
```

```python
X_train
```

```python
# BOW model
from sklearn.feature_extraction.text import CountVectorizer
# for binary BOW enable=Tree
cv = CountVectorizer(max_features = 2500, ngram_range=(1,2))
```

```python
X_train=cv.fit_transform(X_train).toarray()
X_test = cv.transform(X_test).toarray()
```

```python
# just to look printing better
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=10000, formatter=dict(float=lambda x: "%.3g" % x))
X_train
```

```python
cv.vocabulary_
```

```python
# applying model
from sklearn.naive_bayes import MultinomialNB
multinb=MultinomialNB()
```

```python
spam_detect_model=multinb.fit(X_train, y_train)
spam_detect_model
```

```python
y_pred = multinb.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

# Creating TF-IDF model

```python
# train-test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20, random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tv=TfidfVectorizer(max_features=2500, ngram_range=(1,2))
X_train=tv.fit_transform(X_train).toarray()
X_test=tv.transform(X_test).toarray()
```

```python
tv.vocabulary_
```

```python
# applying model
from sklearn.naive_bayes import MultinomialNB
spam_tfidf_model=MultinomialNB().fit(X_train,y_train)
```

```python
# preddictions
y_pred = spam_tfidf_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```
